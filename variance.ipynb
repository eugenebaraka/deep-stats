```python
# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head(7)
```

```python
df.shape
```

```python
# let's plot the distribution of the sepal length

sns.displot(df, x='sepal length (cm)', kind='kde')
plt.title('Distribution of the sepal length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.axvline(df['sepal length (cm)'].mean(), color='r', linestyle='--', label='mean')
plt.legend()
plt.show()
```

```python
df['sepal length (cm)'].describe()
```

The mean of sepal length is 5.84 cm. Steps to calculate variance:
- Calculate the distance of each data point from the mean. 

```python
#putting the plots together

df['distance'] = df['sepal length (cm)'] - df['sepal length (cm)'].mean()

sns.histplot(data=df, x="sepal length (cm)", color="skyblue", label="Sepal Length", kde=True)
sns.histplot(data=df, x="distance", color="red", label="distance from mean", kde=True)
plt.legend()
```

```python
df['distance_squared'] = df['distance'] ** 2
```

```python
sns.histplot(data=df, x="sepal length (cm)", color="skyblue", label="Sepal Length", kde=True)
sns.histplot(data=df, x="distance_squared", color="red", label="distance squared", kde=True)
plt.axvline(df['distance_squared'].mean(), color='b', linestyle='--', label='mean of squared differences')
plt.legend()
```

The mean of squared differences is also known as Variance!! This variance is the average distance between the mean and data points, which represents the overall spread of data. 

```python
print(f"In our case the mean of squared differences or variance is {df['distance_squared'].mean()}")
```

This means that the average distance between each data point and the mean is 0.68 cm. 

```python
def var_calc(x):
    sum_squares = 0
    for num in x:
        sum_squares += (num - np.mean(x))**2
        var = sum_squares/len(x)
    return var
```

```python
var_calc(df['sepal length (cm)'])
```

```python
np.var(df['sepal length (cm)'])
```
